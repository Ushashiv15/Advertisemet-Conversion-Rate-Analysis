```python
#importing pandas package to upload the data set
import pandas as pd
df=pd.read_csv('Merged file.csv')
df
```

```python
#Getting the information about the data set features
df.info()
```

```python
#Deriving the unique values in the given column
df['Est. Clicks'].unique()
```

```python
#Checking the inconsistencies in the data set
df.sum().isnull()
```

```python
#Converting the Data type to numeric
df['Average CPC (USD)']=pd.to_numeric(df['Average CPC (USD)'],errors='coerce')
df['Est. Clicks']=pd.to_numeric(df['Est. Clicks'],errors='coerce')
df['Est. Impressions']=pd.to_numeric(df['Est. Impressions'],errors='coerce')
df['Est. Spend (USD)']=pd.to_numeric(df['Est. Spend (USD)'],errors='coerce')
```

```python
df.info()
```

```python
#Copying the Dataset into control and test Dataframes
A_type=df.copy()
B_type=df.copy()
```

```python
#Increasing the cpc to 1.5
B_type['Average CPC (USD)']=B_type['Average CPC (USD)']*1.5
```

```python
#Calculating the estimated spend 
B_type['Est. Spend (USD)']=B_type['Average CPC (USD)']*B_type['Est. Clicks']
```

```python
#Visualising the calculations of conversion rate and estimated spend in stacked bar chart and line chart
fig,ax=plt.subplots()
b1=ax.bar(es_vs['day'], es_vs['initial_conrate'], color='green',label='init_coversion rate')
b2=ax.bar(es_vs['day'], es_vs['cal_conrate'],bottom=es_vs['initial_conrate'], color='b',label='cal_conversion rate')
l=ax.plot(es_vs['day'], es_vs['Est_spend'], color='red', marker='o', label='Est. Spend')
ax.set_xlabel('DAYS')
ax.set_ylabel('Conversion rate per day')
ax.set_title('Visualisation of the Increased CPC to 1.5 from 1')
ax.legend()
plt.show()
```
